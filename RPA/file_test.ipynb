```python
# 파일 이동
shutil.move(original_dir,target_dir)
```

```python
PDF_DIR = './실습1/pdf_files/'
file_list = os.listdir(PDF_DIR)

for i in range(len(file_list)):
    try:
        dir_name = file_list[i].split('_')
        dir_name_new = dir_name[1]
        os.makedirs(PDF_DIR + dir_name_new)
    except:
        pass    
    target_dir = PDF_DIR + dir_name_new
    original_dir = PDF_DIR + file_list[i]
    shutil.move(original_dir,target_dir)
    print(i)
    i += 1
```
